Nosso 1 AQ: Teletransporte

```python
from qiskit import *
```

```python
circuito = QuantumCircuit(3,3)
```

```python
%matplotlib inline
circuito.draw(output = 'mpl')
```

1) Passo : Sair do estado 0 - > 1, precisamos aplicar a matriz de pauli X

```python
circuito.x(0)
circuito.barrier()
circuito.draw()
```

2 Passo: Vamos usar o segundo qubit como recurso. Come√ßamos aplicando o gate de Hadamard e depois a porta CX

```python
circuito.h(1)
circuito.cx(1,2)


circuito.draw(output = 'mpl')
```

```python
circuito.cx(0,1)
circuito.h(0)
circuito.barrier()
circuito.draw()
```

3) Executar as medidas no qubit 0 e no 1

```python
circuito.measure([0,1],[0,1])
circuito.draw()
```

4) Passo: Vamos aplicar um gate CX do 1 ao 2 e depois um gate CZ.

```python
circuito.barrier()
circuito.cx(1,2)
circuito.cz(0,2)
circuito.draw()
```

```python
circuito.barrier()
circuito.measure(2,2)
circuito.draw()

```

```python
simulador = Aer.get_backend('qasm_simulator')
resultado = execute(circuito, backend = simulador, shots = 1024).result()
counts = resultado.get_counts()
```

```python
from qiskit.tools.visualization import plot_histogram
```

```python
plot_histogram(counts)
```
