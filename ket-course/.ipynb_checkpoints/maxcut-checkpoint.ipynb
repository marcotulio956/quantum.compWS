```python
from ket import *
import pprint as pp
import networkx as nx
from random import uniform
from scipy.optimize import minimize
import matplotlib.pyplot as plt
```

```python
class G:
    def __init__(self, n_nodes, edges, idx=None):
        self.n_nodes = n_nodes
        self.edges = edges
        self.idx = idx
    
    def __iter__(self):
        return G(self.n_nodes, self.edges, -1)
    
    def __next__(self):
        self.idx += 1
        if self.idx < len(self.edges):
            return self.edges[self.idx]
        else:
            raise StopIteration

square = nx.Graph()
square.add_nodes_from([0, 1, 2, 3])
square.add_edges_from([(0, 1), (1, 2), (2, 3), (3, 0)])
options = {
    'node_color': 'red',
    'node_size': 300,
    'width': 3,
}
nx.draw(square, with_labels=True, **options)

g = G(4, [(0, 1), (1, 2), (2, 3), (3, 0)])
```

# Max-Cut Problem

For a graph, a maximum cut is a cut whose size is at least the size of any other cut. That is, it is a partition of the graph's vertices into two complementary sets S and T, such that the number of edges between the set S and the set T is as large as possible. The problem of finding a maximum cut in a graph is known as the Max-Cut Problem.

For the Max-cut we define the objective function C, where the input is a graph with *n* vertices and an edge set ${<ij>}$ of size *m*. The goal is to find a string $z$ that makes 

$$\large C=\sum_{<ij>} C_{<ij>},$$

as large as possible.

# Problem Unitary

We define the problem Hamiltonian

$$\large C=\sum_{<ij>} C_{<ij>} = \sum_{<ij>}\frac{1}{2}(-Z_iZ_j+1).$$ 

From $C$ we define the problem unitary $U_C$ as follows:

$$\Large U_C = e^{-i\gamma C} = \prod_{<ij>} e^{-i\gamma C_{<ij>}}$$

```python
def U_C(gamma, g, q):
    for i, j in g:
        with around(cnot, q[i], q[j]):
            RZ(2*gamma, q[j])
```

# Mixing Unitarie

$$\Large B = \sum_{j=1}^n X_j$$

$$\Large U_B = e^{-i\beta B} = \prod_{j=1}^n e^{-i\beta X_j}$$


```python
def U_B(beta, q):
    RX(2*beta, q)
```

# Initial State

$$\Large \left|s\right> = \left|+ \ldots +\right>$$

```python
def initial_state(q):
    H(q)
```

# QAOA Circuit



```python
def QAOA_base(gammas, betas, g : G, p):
    quatum_circuit = quant(g.n_nodes)
    
    initial_state(quatum_circuit)
    for i in range(p):
        U_C(gammas[i], g, quatum_circuit)
        U_B(betas[i], quatum_circuit)
    
    return quatum_circuit
```

# Expectation Value Estimation

```python
def expval(edge_set):
    val = 0.0
    for edge in edge_set.states:
        # If edges belong to the same cut decrease value
        # else increase value
        if bool(edge & 1) != bool(edge & 2):
            val -= edge_set.probability(edge)
        else:
            val += edge_set.probability(edge)
    return val

def QAOA(gammas, betas, g : G, p):
    # Creating Quantum Circuit
    quantum_circuit = QAOA_base(gammas, betas, g, p)

    # Measuring states
    measured_states = [dump(quantum_circuit[j]+quantum_circuit[k]) for j, k in g]

    # Return C = sum Cjk = sum 1/2(-ZjZk + 1)
    return sum(0.5 * (1 - expval(state)) for state in measured_states) 
    
```

# QAOA applied to MaxCut

```python
def maxcut(g : G, p=1):
    # Objective Function to be optimized
    def objective(params):
        gammas = params[:p]
        betas = params[p:]
        # Since we are minimizing the problem we apply
        # a minus sign to the result of QAOA
        return -QAOA(gammas, betas, g, p)

    # Defining how many times we are repeating the
    # optimization algorithm
    iterations = 1
    
    all_results = []
    for i in range(iterations):
        # Initializing 2p parameters gamma and beta
        params = [uniform(0, 1) for _ in range(p*2)]
        
        # Optimizing QAOA using classical optimization algorithm
        res = minimize(objective, params, method='COBYLA')
        print(res)
        
        # Saving results
        all_results.append((res.x, res.fun))

    # Return best result from the distribution
    all_results.sort(key=lambda tup: tup[1]) 
    params = all_results[0][0]
    
    # Plotting answer
    answer = dump(QAOA_base(params[:p], params[p:], g, p))
    _, ax = plt.subplots()
    ax.bar([f"{i:0{g.n_nodes}b}" for i in answer.states if answer.probability(i) > 1e-6], [answer.probability(i) for i in answer.states if answer.probability(i) > 1e-6])
    plt.xticks(rotation=90)
    #plt.show()
```

```python
maxcut(g, 1)
```
